#Playing Around w/ Convolutional Filters and Edge Detection

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
import numpy as np

%matplotlib inline
```

## Load image

```python
xray_bgr = cv2.imread('resources\\filter_images\\chest_xray.jpg')
xray_rgb = cv2.cvtColor(xray_bgr, cv2.COLOR_BGR2RGB)

plt.imshow(xray_rgb)
```

## Apply a custom filter, with the hope of highlighting hotspots

```python
custom_kernel1 = np.array([[ -3, 0, -3],
                        [0, 12, 0],
                        [-3, 0, -3]])

custom_image = cv2.filter2D(xray_rgb, -1, custom_kernel1)
plt.imshow(custom_image)
```

## Seeing if hotspot can be accentuated by increasing the dims of the filter

```python
custom_kernel2 = np.array([[ -5, -3, 0, -3, -5],
                        [-3, 0, 38, 0, -3],
                        [-5, -3, 0, -3, -5]])

custom_image2 = cv2.filter2D(xray_rgb, -1, custom_kernel2)
plt.imshow(custom_image2)
```
