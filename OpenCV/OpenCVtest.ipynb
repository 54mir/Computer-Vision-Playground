```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread('bluetree.jpg')


print("Type: ", type(image))
print("Dimensions: ", image.shape)
```

```python
%matplotlib inline

# Make Copy of Image
imageCopy = np.copy(image)

# Convert to RGB
imageCopy = cv2.cvtColor(imageCopy, cv2.COLOR_BGR2RGB)

plt.imshow(imageCopy)
```

# Define Color threshold and mask

```python
# Define Color Threshold
lowerBlue = np.array([0, 0, 230])
upperBlue = np.array([50, 50, 255])

# Create mask
mask = cv2.inRange(imageCopy, lowerBlue, upperBlue)
maskCopy = np.copy(mask)

plt.imshow(mask, cmap="gray")
```
