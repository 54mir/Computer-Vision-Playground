```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread('resources/bluetree.jpg')


print("Type: ", type(image))
print("Dimensions: ", image.shape)
```

```python
%matplotlib inline

# Make Copy of Image
imageCopy = np.copy(image)

# Convert to RGB
imageCopy = cv2.cvtColor(imageCopy, cv2.COLOR_BGR2RGB)

plt.imshow(imageCopy)
```

# Define Color threshold and mask

```python
# Define Color Threshold
lowerBlue = np.array([0, 0, 100])
upperBlue = np.array([80, 120, 255])

# Create mask
mask = cv2.inRange(imageCopy, lowerBlue, upperBlue)


plt.imshow(mask, cmap="gray")
```

```python
# Copy original image, hide the white area of the mask
maskedImage = np.copy(imageCopy)
maskedImage[mask != 0] = [0, 0, 0]
plt.imshow(maskedImage)
```

```python
# Load in a background image, convert to RGB, and crop to foreground image size
background = cv2.imread('resources/mars.jpg')
cv2.cvtColor(background, cv2.COLOR_BGR2RGB)
background = background[0:720, 0:1280]
```

```python
# Apply mask to background (i.e. hide black area of mask)

maskedBackground = np.copy(background)
maskedBackground[mask == 0] = [0, 0, 0]
plt.imshow(maskedBackground)
```


## Complete Image

```python
#Complete image
completeImage = maskedImage + maskedBackground
plt.imshow(completeImage)
```
